```python
# Write code which calculates the area of a cylinder by taking user input for radius and height
# pi is math.pi from the math library
import math

radius = float(input("What is the radius of the cylinder?"))
height = float(input("What is the height of the cylinder?"))
print(math.pi * radius ** 2 * height)
```

```python
# switching datatypes
int1 = 7
float1 = 4.35
str1 = "8.20"
str2 = "8"
print(str(int1))
print(str(float1) + "\n")

print(int(float1))
print(int(str2), "\n")
# print(int(str1))

print(float(int1))
print(float(str1))
print(float(str2))
```

```python
# String slicing 

str1 = "I was wondering why the frisbee kept getting bigger and bigger, but then it hit me."

print(str1[3:9])
```
